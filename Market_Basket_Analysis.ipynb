```python
# Create list of lists for encoding
trans=[]
for i in range (0,7501):
    trans.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
# Prepare data for Apriori function
TE = TransactionEncoder()
array = TE.fit(trans).transform(trans)
```

```python
# Convert data set back to dataframe 
cleaned_df = pd.DataFrame(array, columns = TE.columns_)
cleaned_df
```

```python
for col in cleaned_df.columns:
    print(col)
```

```python
df_cleaned = cleaned_df.drop(['nan'], axis=1)
```

```python
for col in df_cleaned.columns:
    print(col)
```

```python
# Export prepared data sets
df.to_csv(r"C:\Users\mlaws\OneDrive - Western Governors University\Documents\WGU\D212\medical_market_basket_prepared.csv")
```

```python
# Run Apriori to mine
a_rules = apriori(df_cleaned, min_support = 0.005, use_colnames = True)
# Review head
a_rules.head(3)
```

```python
# Set metric
ass_r = association_rules(a_rules, metric = 'lift', min_threshold = 1)
ass_r.head(3)
```

```python
df_cleaned['abilify'].sum()
```

```python
# Generate top 3 rules 
ass_r[ (ass_r['lift'] >= 6.3) & ass_r['confidence'] > 0.2][:3]
```
